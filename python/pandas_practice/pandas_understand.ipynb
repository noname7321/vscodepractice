```python
people = {
    'first':['John', 'Jane', 'Jim'],
    'last':['Doe', 'Smith', 'Beam'],
    'email':['JohnDoe@email.com', 'JaneSmith@email.com', 'JimBeam@email.com']
}
```

```python
people['email']
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(people)
df
```

```python
df['email'] # preferred way to access a column
```

```python
type(df['email'])
```

```python
df.email # there is a risk if column name is same as DataFrame method or attribute
```

```python
df[['first', 'last']] # passing a list to access multiple columns
```

```python
df.columns # to see all column names
```

```python
df.iloc[0]  # access first row by position, returns a Series
```

```python
df.iloc[[0,2]]  # access first and third rows by position, returns a DataFrame
```

```python
df.iloc[[0,2],2]  # access first and third rows, third column (email)
```

```python
print(df.loc[0]) # access first row by label, returns a Series
print('\n')  
print(df.loc[0,'email'])  # access first row, email column
print('\n')
df.loc[[0,1]]
```

```python
df.loc[[0,1],['email','last']]  # access first and second rows, email and last columns
```
