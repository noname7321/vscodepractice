```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first':['John', 'Jane', 'Jim'],
    'last':['Doe', 'Smith', 'Beam'],
    'email':['JohnDoe@email.com', 'JaneSmith@email.com', 'JimBeam@email.com']
}
```

```python
df=pd.DataFrame(people)
df.set_index('email')# sets email as index without modifying original DataFrame
```

```python
df
```

```python
df.set_index('email', inplace=True) # modifies the original DataFrame
df
```

```python
df.index
```

```python
#print(df.loc['0']) # This will raise an error since the index is now email
print(df.iloc[0]) # Accessing by integer location still works
print('\n')
print(df.loc['JohnDoe@email.com'])
print('\n')
print(df.loc['JohnDoe@email.com', 'first'])
```

```python
df.reset_index(inplace=True)  # Resets index to default integer index
df
```
