原因: 你的 df.index 是 DatetimeIndex 但 is_monotonic_increasing=False（非单调），Pandas 不允许在非单调的 DatetimeIndex 上做基于标签的部分字符串切片（尤其当边界键不存在时），因此抛出 KeyError。

快速修复（推荐）：

确保 Date 为时间类型：
df['Date'] = pd.to_datetime(df['Date'])

将 Date 设为索引：
df = df.set_index('Date')

对索引排序使其单调递增：
df = df.sort_index() 或 df.sort_index(inplace=True)

再执行部分切片：
df['2020-01-01 00:00:00':'2020-12-31 23:59:59']

可以利用以下代码排查：

print(type(df.index))

print(df.index.dtype)

print(df.index.is_monotonic_increasing)

```python
df['2020-01-01 00:00:00':'2020-12-31 23:59:59']
```

```python
df['2020-01-01 00:00:00':'2020-12-31 23:59:59']['Close'].mean()
```

df['2020-01-01'] 这个写法对 DataFrame 来说优先当作按列取值（去找列名为 '2020-01-01'），因此当没有该列时会抛 KeyError。只有用 .loc[...] 或显式切片（在索引为单调的 DatetimeIndex 时）才会按行（索引）进行时间点/时间段选择。
