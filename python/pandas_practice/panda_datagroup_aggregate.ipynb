```python
country_people_who_use_python = country_grp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
countryrespondent=df['Country'].value_counts()
python_df=pd.concat([countryrespondent, country_people_who_use_python], axis=1, sort=False)
python_df.columns=['TotalRespondent', 'PythonUsers']
python_df['PercentPythonUsers'] = python_df['PythonUsers']/python_df['TotalRespondent']*100
python_df.sort_values(by='PercentPythonUsers', ascending=False, inplace=True)
python_df.head(50)
```
