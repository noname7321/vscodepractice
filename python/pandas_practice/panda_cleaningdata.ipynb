```python
import pandas as pd
import numpy as np
```

```python
people = {
    'First': ['Alice', 'Bob', 'Charlie', 'David', 'Eva',np.nan,None,'NA', 'Frank', 'Grace', 'Hannah'],
    'Last': ['Smith', 'Jones', 'Brown', 'Missing', None, 'Davis', 'Miller', 'Wilson', 'Moore', 'Taylor', 'Anderson'],
    'Age': [25, 30, np.nan, 35, None, 28, 40, 22, 29, np.nan, 31],
    'City': ['New York', 'Los Angeles', 'New York', None, 'Chicago', 'NA', 'Chicago', 'Los Angeles', 'New York', 'Chicago', 'Los Angeles']
}
df= pd.DataFrame(people)
df.replace(['NA', 'Missing'], np.nan, inplace=True)
```

```python
df.isna() # df.fillna('missing')
```

```python
print(type(np.nan))
df.dtypes
```

```python
df['Age']=df['Age'].astype('float') 
# to change datatype of 'Age' column to float,can't convert to int as there are NaN values
```

```python
df['Age'].mean()
```

```python
df.dropna(axis=0, how='any')  # how='any' or 'all','any' means if any NA value is present in row/column, drop that row/column
# 'all' means drop row/column only if all values are NA
```

```python
df.dropna(axis='index', how='any',subset=['Last','First'])  # drop rows where 'Last' or 'First' is NA
```
