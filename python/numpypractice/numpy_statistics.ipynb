Statistics

```python
import numpy as np
```

```python
stats = np.array([[10, 200, 300], [5, 2, 250], [1, 150, 3], [100, 120, 130]])
print(stats)
```

```python
print(np.min(stats))
print(np.min(stats, axis=0))# find the minimum value in each column
print(np.min(stats, axis=1))# find the minimum value in each row
print(np.max(stats))
print(np.mean(stats))
print(np.median(stats))
print(np.std(stats))
print(np.sum(stats))
```
